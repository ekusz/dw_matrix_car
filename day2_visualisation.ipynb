```
 def gruop_and_barplot(feat_groupby, feat_agg='price_value', agg_fundcs=[np.mean, np.median, np.size], feat_sort='mean', top=50, subplots=True):
  return (
    df
    .groupby(feat_groupby)[feat_agg]
    .agg(agg_fundcs)
    .sort_values(by='size', ascending=False)
    .head(top)
  
).plot(kind='bar' , figsize=(15,5), subplots=subplots)
```

```
gruop_and_barplot('param_marka-pojazdu');
```

```
gruop_and_barplot('param_kraj-pochodzenia', feat_sort='size');
```

```
gruop_and_barplot('param_kolor', feat_sort='mean');
```
